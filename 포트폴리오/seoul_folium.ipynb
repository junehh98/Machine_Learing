```python
import pandas as pd
df_dash1 = pd.read_excel('/content/서울시 대기질 자료 제공_2016-2019.xlsx',na_values='..')
df_dash2 = pd.read_excel('/content/서울시 대기질 자료 제공_2020-2021.xlsx',na_values='..')
```

```python
df_dash = pd.concat([df_dash1, df_dash2])
```

```python
df_newdash =df_dash.drop(['ultrafine dust'],axis = 1)  #초미세먼지 컬럼 삭제
```

```python
df_newdash['구분'].unique() # 서울시 행정구역 별로 출력
```

```python
agg = ['평균', '강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구',
       '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구',
       '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']
```

```python
df_agg = df_newdash[df_newdash['구분'].isin(agg)]
```

```python
df_nonagg = df_newdash[df_newdash['구분'].isin(agg)]
df_nonagg.head() # 포함 안되것 없음 확인
```

```python
df_flat = df_nonagg.melt(id_vars=["일시","구분"],   #데이터 형태 wide to long
    var_name = "dust",
    value_name = "value")

df_flat.head(),df_flat.tail()
```

```python
df_flat.shape
```

```python
print( df_flat.shape ) 
df_flat.head()
```

```python
df_flat[['dust','NA']] = df_flat.dust.str.split(" ", expand=True) #공백 기준으로 컬럼 분리
import datetime as dt
df_flat['일시'] = df_flat['일시'].dt.strftime('%d%m%Y') # 일시 값을 object로 변경
df_flat.dtypes
```

```python
df_flat['일시'] = pd.to_numeric(df_flat['일시'])
df_flat=df_flat.dropna(axis=0, subset = ['value']) #결측치 행 제거
```

```python
df_flat
```

```python
df_flat = df_flat.groupby(['일시', '구분'], as_index=False).mean()
df_flat
```

```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
print(folium.__version__)
```

```python
m = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11) # 서울시 좌표 설정
```

```python
state_geo = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
# github에서 서울지도 자료 불러오기
```

```python
m.choropleth(
    geo_data=state_geo,
    name='dust',
    data=df_flat,
    columns=['구분','value'],
    key_on='feature.properties.name',
    fill_color='Reds',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = 'income'
)
```
